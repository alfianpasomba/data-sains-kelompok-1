```python
import pandas as pd
import seaborn as sns
sns.set_style("darkgrid")
```

## Hannan (Line Chart)

```python
df1 = pd.read_csv("Data Komoditas.csv")
df1.head()
```

```python
linechart = sns.lineplot(data=df1, x="tahun", y="nilai", hue="kategori")
linechart.set_title("Linechart Komoditas Teh")
```

## Kahfi (Bar Chart)

```python
df2 = pd.read_csv("Data Penduduk Jakarta.csv")
df2.head()
```

```python
barplot = sns.barplot(data=df2, x="Laki-laki (2020)", y="Kelompok Umur")
barplot.set_title("Barchart Penduduk Jakarta Laki-laki 2020")
```

## Evi (Histogram)

```python
df3 = pd.read_csv("Data Luas Wilayah.csv")
df3.head()
```

```python
histogram = sns.histplot(data=df3, x="2019")
histogram.set_title("Histogram Luas Wilayah 2019")
```

## Yasrul (Scatter Plot)

```python
df4 = pd.read_csv("Data Penduduk Sulsel.csv")
df4.head()
```

```python
scatterplot = sns.scatterplot(data=df4, x="2021", y="Wilayah")
scatterplot.set_title("Scatterplot Jumlah Penduduk Sulsel 2021")
```

## Alfian (Boxplot)

```python
df5 = pd.read_csv("Data Mahasiswa.csv")
df5.head()
```

```python
boxplot = sns.boxplot(data=df5, x="Akreditasi", y="Jumlah Mahasiswa")
boxplot.set_title("Boxplot Jumlah Mahasiswa Berdasarkan Akreditasi")
```

## Maulana (Heat Map)

```python
df6 = pd.read_csv("Data Kemiskinan.csv",index_col="Kabupaten")
df6.head()
```

```python
heatmap = sns.heatmap(df6,yticklabels=True)
heatmap.set_title("Heatmap Jumlah Penduduk Miskin")
```
